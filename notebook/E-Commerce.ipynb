```python
# Annahme: Der DataFrame heißt 'df3'.

# Konvertiere die Spalte 'Datetime' in das richtige Datumsformat.
df3['Datetime'] = pd.to_datetime(df3['Datetime'], format='%d.%m.%Y')

# Erstelle eine neue Spalte für das Quartal.
df3['Quarter'] = df3['Datetime'].dt.quarter

# Gruppiere den DataFrame nach 'sub-category' und 'Quarter' und erhalte die Zeile mit den meisten 'clicks' für jede Gruppe.
best_quarter_per_category = df3.groupby(['sub-category', 'Quarter']).apply(lambda x: x.nlargest(1, 'clicks')).reset_index(drop=True)

# Gruppiere den DataFrame nach 'sub-category' und berechne den Gesamtanteil der 'clicks' für jede 'sub-category'.
total_clicks_per_category = df3.groupby('sub-category')['clicks'].sum().reset_index()

# Füge die Gesamtanteilsspalte in den DataFrame mit den besten Quartalsdaten ein.
best_quarter_per_category = best_quarter_per_category.merge(total_clicks_per_category, on='sub-category', suffixes=('_Best_Quarter', '_Total'))

# Berechne den Anteil der 'clicks' am Gesamtwert für jede 'sub-category' im besten Quartal.
best_quarter_per_category['Anteil_Klicks_an_Gesamt'] = best_quarter_per_category['clicks_Best_Quarter'] / best_quarter_per_category['clicks_Total']

# Gib den neuen DataFrame mit den gewünschten Ergebnissen aus.
print(best_quarter_per_category[['sub-category', 'Quarter', 'Anteil_Klicks_an_Gesamt']])
```
